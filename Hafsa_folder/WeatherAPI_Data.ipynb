```python
import requests as re
import pandas as pd
import os
from IPython.display import JSON
import json
```

### Chicago

```python
import datetime
url = 'https://api.worldweatheronline.com/premium/v1/past-weather.ashx'

# date in yyyy/mm/dd format
enddate1 = datetime.datetime(2020, 6, 30)
startdate = datetime.datetime(2018, 1, 1)
str_date = '2018-01-01'
count = 0

params_1 = dict(
q='Chicago,il',
date=str_date,
enddate='2020-06-30',
tp='12',
format='json',
key='e7a4e5710627475fafa54006210703'
)

my_dict = {'dates_list': [],
            'temp_night_list' : [],
            'temp_day_list': [],
            'speed_night' : [],
            'speed_day' : [],
             'desc_night': [],
           'desc_day': []
          }


while startdate < enddate1:
    params_1 = dict(
        q='Chicago,il',
        date=str_date,
        enddate='2020-06-30',
        tp='12',
        format='json',
        key='e7a4e5710627475fafa54006210703'
        )
    resp_1 = re.get(url=url, params=params_1)
    data_1 = json.loads(resp_1.text)
    JSON(data_1)
    
    str_date = data_1['data']['weather'][-1]['date']
    
    startdate = datetime.datetime.strptime(str_date, '%Y-%m-%d')
    
    for i in data_1['data']['weather']:
        
        date = i['date']
        my_dict['dates_list'].append(date)
        
        temp_night = i['hourly'][0]['tempC']
        my_dict['temp_night_list'].append(temp_night)
        
        temp_day = i['hourly'][1]['tempC']
        my_dict['temp_day_list'].append(temp_day)
        
        speed_night = i['hourly'][0]['windspeedKmph']
        my_dict['speed_night'].append(speed_night)
        
        speed_day = i['hourly'][1]['windspeedKmph']
        my_dict['speed_day'].append(speed_day)
        
        desc_night = i['hourly'][0]['weatherDesc'][0]['value']
        my_dict['desc_night'].append(desc_night)
        
        desc_day = i['hourly'][1]['weatherDesc'][0]['value']
        my_dict['desc_day'].append(desc_day)
```

### Chicago dataframe

```python
df_chicago = pd.DataFrame(my_dict)
```

```python
df_chicago.to_csv('chicago.csv', index=False)
```
