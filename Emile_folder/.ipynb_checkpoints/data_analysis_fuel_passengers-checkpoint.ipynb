## Task 2
add passengers per month per carrier

```python
df_passenger = pd.read_csv('/home/emile/Documents/data_bootcamp/w6/passengers_orig.csv')
```

```python
delta = df_passenger['carrier'] == 'DL'
df_passenger[delta]  # looking at specific carrier
```

```python
df_passenger[delta]['year'].unique() # data for delta carrier has 5 years, but might not have data for every month of each year
```

```python
df_passenger[delta].groupby(by='month')['passengers'].sum()/len(df_passenger[delta]['year'].unique()) # possible way to get passengers per month on average?
```

```python
delta_month = (df_passenger['carrier'] == 'DL') & (df_passenger['month'] == 7) & (df_passenger['year'] == 2019)
```

```python
df_passenger[delta_month]
```

```python
_2019 = df_passenger['year'] == 2019 
passengers_2019 = df_passenger[_2019]
passengers_2019 # might take passengers per month for 2019 and use that
```

```python
list_relevant_carriers = ['WN', 'DL', 'MQ', 'YV', 'PT', 'UA', 'OO', 'EV', 'G7', 'YX', '9E',
                         'G4', 'AA', 'CP', 'QX', 'NK', 'AX', 'HA', 'B6', 'OH', 'AS', 'F9',
                         'ZW', 'C5', 'EM', 'KS', 'VX']
```

```python
passengers_2019['unique_carrier'].unique()
```

```python
relevant_carriers_pass = passengers_2019[passengers_2019['unique_carrier'].isin(list_relevant_carriers)]
```

```python
relevant_carriers_pass
```

```python
pass_per_month_carrier = relevant_carriers_pass.groupby(by=['unique_carrier', 'month'])['passengers'].sum().reset_index()
```

```python
pass_per_month_carrier.sort_values(by='passengers', ascending=False)
```

```python
pass_per_month_carrier = pass_per_month_carrier.rename(columns={'unique_carrier': 'op_unique_carrier'})
```

```python
new_df2 = pd.merge(new_df, pass_per_month_carrier , on=['op_unique_carrier', 'month'], how='outer', indicator=True)
```

```python
new_df2
```

```python
left_only = new_df2['_merge'] == 'left_only'
new_df2[left_only]
```

```python
new_df2['passengers']=new_df2.groupby('month')['passengers'].apply(lambda x:x.fillna(x.mean()))
```

```python
new_df2[left_only]
```

```python
new_df2.isnull().sum()
```

```python
new_df2 = new_df2.dropna()
```

```python
new_df2 = new_df2.drop(labels='_merge', axis=1)
```

```python
new_df2 = new_df2.rename(columns={'passengers': 'monthly_passenger_carrier'})
```

```python
new_df2
```

## Task 3
add passengers per month per airport

```python
monthly_passenger_airport = df_passenger.groupby(by=['origin_city_name', 'month'])['passengers'].sum().reset_index()
```

```python
monthly_passenger_airport
```

```python
unique_cities = new_df2['origin_city_name'].unique()
relevant_airport_monthly_pass = monthly_passenger_airport[monthly_passenger_airport['origin_city_name'].isin(unique_cities)]
```

```python
relevant_airport_monthly_pass
```

```python
new_df3 = pd.merge(new_df2, relevant_airport_monthly_pass , on=['origin_city_name', 'month'], how='outer', indicator=True)
```

```python
new_df3
```

```python
left_only2 = new_df3['_merge'] == 'left_only'
new_df3[left_only2]
```

```python
new_df3['passengers']=new_df3.groupby('month')['passengers'].apply(lambda x:x.fillna(x.mean()))
```

```python
new_df3 = new_df3.dropna()
```

```python
new_df3 = new_df3.rename(columns={'passengers':'monthly_passenger_airport'})
new_df3 = new_df3.drop(labels='_merge', axis=1)
```
