# DF for flights

```python
df = pd.read_csv('/home/emile/Documents/data_bootcamp/w6/df_flights_orig.csv')
```

```python
df.drop(['no_name', 'first_dep_time', 'total_add_gtime', 'longest_add_gtime', 'cancellation_code'], axis=1, inplace=True)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df['tail_num']=df['tail_num'].fillna('Unknown')    # add unknown to tail numbers if not there
```

```python
dep_nan = df['dep_time'].isnull()
df[dep_nan]
```

```python
for index, row in df.iterrows():  # get rid of cancelled flights
    if row['cancelled'] == 1.0:
        df.drop(index=index, inplace=True)
```

```python
arr_time = df['arr_time'].isnull()   # no arrival time and not cancelled but there's a depart time. crashed?
df[arr_time]
```

```python
df['arr_delay'] = df['arr_delay'].fillna('none')
for index, row in df.iterrows():   # drop rows missing target variable
    if row['arr_delay'] == 'none':
        df.drop(index=index, inplace=True)
```

```python
df['wheels_off'] = df['wheels_off'].fillna('none')
for index, row in df.iterrows():   # dropping rows missing datetime information
    if row['wheels_off'] == 'none':
        df.drop(index=index, inplace=True)
```

```python
df.isnull().sum()
```

```python
df['dep_delay'].fillna(0, inplace=True)   # no difference in scheduled and departure time, so 0 delay
```

```python
df['carrier_delay'] = df['carrier_delay'].fillna(0.0)
df['weather_delay'] = df['weather_delay'].fillna(0.0)
df['nas_delay'] = df['nas_delay'].fillna(0.0)
df['security_delay'] = df['security_delay'].fillna(0.0)
df['late_aircraft_delay'] = df['late_aircraft_delay'].fillna(0.0)
```

```python
df.isnull().sum()
```

```python
df.head()
```

```python
df.to_csv(r'./df_cleaned.csv', index = False)
```
