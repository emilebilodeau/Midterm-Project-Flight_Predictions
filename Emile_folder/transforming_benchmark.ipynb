# Grid Search

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid1 = {
    'alpha': [0.001, 0.01, 0.1, 0.25, 0.5, 0.75, 1, 3, 5]
}

param_grid2 = {
    'alpha': [0.001, 0.01, 0.1, 0.25, 0.5, 0.75, 1, 3, 5],
    'l1_ratio': [0, 0.25, 0.5, 0.75, 1]
}

lasso = linear_model.Lasso()
ridge = linear_model.Ridge()
elastic = linear_model.ElasticNet()
```

## Grid search lasso

```python
grid = GridSearchCV(estimator=lasso, param_grid=param_grid1, scoring='r2', cv=100, verbose=1, n_jobs=-1)
```

```python
grid_result = grid.fit(X_train_scaled, y_train)
```

```python
print(grid_result.best_score_)
print(grid_result.best_params_)
```

```python
y_pred_g1 = grid.predict(X_test_scaled)
```

```python
y_pred_g1
```

```python
g1_MSE = mean_squared_error(y_test, y_pred_g1)
print(g1_MSE)
```
